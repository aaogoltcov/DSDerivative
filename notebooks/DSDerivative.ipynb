```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
import pandas as pd
from sympy import diff, symbols, cos, sin
```

```python
# Уровень 0
# посчитайте значение 1-ой и 2-ой производной фунции
# f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1
```

```python
x, y = symbols('x y')
def f(x):
    return pow(x, 5) + 4 * sin(2 * x) + cos(3 * x + 3)
    
def derivative_first(function):
    return diff(function)

def derivative_second(derivative_first):
    return diff(derivative_first)

display(f(x))
display(derivative_first(f(x)))
display(derivative_second(derivative_first(f(x))))

print(f(1))
print('1st derivative: ', derivative_first(f(1)), '-- another method: ', derivative(f, 1, dx=1e-6, n=1))
print('2nd derivative: ', derivative_second(derivative_first(f(1))), '-- another method: ', derivative(f, 1, dx=1e-6, n=2))

xf = np.linspace(-10, 10, 20)
yf = [f(xf[i]) for i in range(len(xf))]
ydf1 = [derivative_first(f(xf[i])) for i in range(len(xf))]
ydf2 = [derivative_second(derivative_first(f(xf[i]))) for i in range(len(xf))]
ydf1_derivative = [derivative(f, xf[i], dx=1e-6, n=1) for i in range(len(xf))]
ydf2_derivative = [derivative(f, xf[i], dx=1e-6, n=2) for i in range(len(xf))]
plt.plot(xf, yf, label='function')
plt.plot(xf, ydf1, label='derivative first')
plt.plot(xf, ydf2, label='derivative second')
plt.plot(xf, ydf1_derivative, label='derivative first (another method)')
plt.plot(xf, ydf2_derivative, label='derivative second (another method)')
plt.legend()
plt.show()
```
